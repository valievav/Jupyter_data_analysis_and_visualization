```python
import folium
import pandas as pd

# prepare data
data = pd.read_html("https://www.latlong.net/category/ports-236-48.html")
df = data[0]

site_lat = df['Latitude']
site_lon = df['Longitude']
locations = df['Place Name']

# create map
world_map = folium.Map(location=[10, -90], tile='cartodbpositron', zoom_start=3)

# place markers
for lat, lon, locat in zip(site_lat, site_lon, locations):
    folium.CircleMarker([lat, lon],
                        radius=3,
                        tooltip = ('<strong>Port</strong>: ' + str(locat).capitalize()),
                        color='red',
                        fill_color='red',
                        fill_opacity=0.7).add_to(world_map)

# add title
title_html = '''
             <h5 align="center" style="font-size:20px"><b>Ports in United States</b></h5>
             '''
world_map.get_root().html.add_child(folium.Element(title_html))

# save into file and display result
world_map.save("Folium_map.html")
display(world_map)
```
